```python
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# WebDriver 초기화
driver = webdriver.Chrome()
driver.implicitly_wait(3)
url = 'https://www.busanbank.co.kr/ib20/mnu/BHPDRC007004000'
driver.get(url)

# WebDriverWait 객체 생성
wait = WebDriverWait(driver, 10)

# 결과 저장 리스트
results_list = []

# SIGUNGU와 Btnm 선택 반복
for i in range(1, 17):
    for j in range(1, 13):
        # 드롭다운 및 검색 작업
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#text_SIGUNGU'))).click()
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'#listSigungu > div > div > div > ul > li:nth-child({i})'))).click()
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'#listBtnm > div > div > div > ul > li:nth-child({j})'))).click()
        wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="btnSearch"]'))).click()

        # 검색 결과 대기 및 수집
        time.sleep(2)  # 검색 결과 로드 대기
        results = driver.find_elements(By.CSS_SELECTOR, '#resultBox > div')
        num_results = len(results)

        # 결과 파싱
        for k in range(1, num_results + 1):
            field = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child{k} > ul > li')
            title = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_title').get_attribute('textContent').strip()
            addr = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_addr').get_attribute('textContent').strip()
            tel = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_tel').get_attribute('textContent').strip()

            result_data = {
                "filed": field,
                "title": title,
                "address": addr,
                "telephone": tel
            }
            results_list.append(result_data)

        # **페이지 새로고침**
        driver.refresh()  # 새로고침
        time.sleep(3)  # 새로고침 후 페이지 로드 대기

        # 새로고침 후 다시 페이지로 이동
        #driver.get(url)
        #time.sleep(3)  #페이지 재로딩 대기

# 결과 JSON 파일 저장
with open('results.json', 'w', encoding='utf-8') as json_file:
    json.dump(results_list, json_file, ensure_ascii=False, indent=4)

driver.quit()

print("Data has been saved to 'results.json'")
```

```python
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

# 헤드리스 모드 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 헤드리스 모드 활성화
chrome_options.add_argument("--disable-gpu")  # GPU 비활성화 (Windows에서 권장)
chrome_options.add_argument("--no-sandbox")  # Linux에서 권장
chrome_options.add_argument("--disable-dev-shm-usage")  # 메모리 사용 제한 해제

# WebDriver 초기화
driver = webdriver.Chrome(options=chrome_options)
driver.implicitly_wait(3)
url = 'https://www.busanbank.co.kr/ib20/mnu/BHPDRC007004000'
driver.get(url)

# WebDriverWait 객체 생성
wait = WebDriverWait(driver, 10)

# 결과 저장 리스트
results_list = []

# SIGUNGU와 Btnm 선택 반복
for i in range(1, 17):
    for j in range(1, 13):
        # 드롭다운 및 검색 작업
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#text_SIGUNGU'))).click()
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'#listSigungu > div > div > div > ul > li:nth-child({i})'))).click()
        wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f'#listBtnm > div > div > div > ul > li:nth-child({j})'))).click()
        wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="btnSearch"]'))).click()

        # 검색 결과 대기 및 수집
        time.sleep(2)  # 검색 결과 로드 대기
        results = driver.find_elements(By.CSS_SELECTOR, '#resultBox > div')
        num_results = len(results)

        # 결과 파싱
        for k in range(1, num_results + 1):
            field = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > ul > li')
            title = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_title').get_attribute('textContent').strip()
            addr = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_addr').get_attribute('textContent').strip()
            tel = driver.find_element(By.CSS_SELECTOR, f'#resultBox > div:nth-child({k}) > p.r_tel').get_attribute('textContent').strip()

            result_data = {
                "filed": field,
                "title": title,
                "address": addr,
                "telephone": tel
            }
            results_list.append(result_data)

        # **페이지 새로고침**
        driver.refresh()  # 새로고침
        time.sleep(3)  # 새로고침 후 페이지 로드 대기

# 결과 JSON 파일 저장
with open('results.json', 'w', encoding='utf-8') as json_file:
    json.dump(results_list, json_file, ensure_ascii=False, indent=4)

driver.quit()

print("Data has been saved to 'results.json'")
```
